```python
chroma_rms = []
chroma_stft =[]
spectral_centroid = []
spectral_bandwidth = []
rolloff = []
zero_crossing_rate = []
harmony = [] 
perceptr = []
tempo = []
mfcc = []

save_path = '__file_location__'

sr = 22050
for mood in tqdm(mood_list):
  filenames = os.listdir(music_path + mood)
  for music_name in filenames:
    loaded_m, sr = librosa.load(music_path + mood + '/' + music_name)
    trim_m, _ = librosa.effects.trim(loaded_m)
    early_part = trim_m[:30*22050]

    chroma_rms.append(librosa.feature.rms(early_part))
    chroma_stft.append(librosa.feature.chroma_stft(early_part))
    spectral_centroid.append(librosa.feature.spectral_centroid(early_part, sr = sr))
    spectral_bandwidth.append(librosa.feature.spectral_bandwidth(early_part, sr = sr)) 
    rolloff.append(librosa.feature.spectral_rolloff(early_part, sr = sr))
    zero_crossing_rate.append(librosa.feature.zero_crossing_rate(early_part))
    
    early_part_harm, early_part_perc = librosa.effects.hpss(early_part)
    harmony.append(early_part_harm)
    perceptr.append(early_part_perc)

    tempo_m, _ = librosa.beat.beat_track(early_part, sr = sr)
    tempo.append(tempo_m)

    mfcc_m = librosa.feature.mfcc(early_part)
    mfcc.append(mfcc_m)

np.save(save_path + 'chroma_rms.npy', chroma_rms)
np.save(save_path + 'chroma_stft.npy', chroma_stft)
np.save(save_path + 'spectral_centroid.npy', spectral_centroid)
np.save(save_path + 'spectral_bandwidth.npy', spectral_bandwidth)
np.save(save_path + 'rolloff.npy', rolloff)
np.save(save_path + 'zero_crossing_rate.npy', zero_crossing_rate)
np.save(save_path + 'harmony.npy', harmony)
np.save(save_path + 'perceptr.npy', perceptr)
np.save(save_path + 'tempo.npy', tempo)
np.save(save_path + 'mfcc.npy', mfcc)
```
